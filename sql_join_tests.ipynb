### INNER JOIN 

returns the records that exist in both tables

```
df_3 = _deepnote_execute_sql('select * \nfrom employee as e\ninner join department as d on e.dep_id=d.dep_id', 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_3
```

```
df_4 = _deepnote_execute_sql('select * \nfrom department as d\ninner join employee as e on d.dep_id=e.dep_id', 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_4
```

### LEFT JOIN 

returns records from the left table, even if there are no such records in the right table

```
df_5 = _deepnote_execute_sql('select * \nfrom employee as e\nleft join department as d on e.dep_id=d.dep_id', 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_5
```

```
df_6 = _deepnote_execute_sql('select * \nfrom department as d\nleft join employee as e on d.dep_id=e.dep_id', 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_6
```

### RIGHT JOIN

returns records from the right table, even if there are no such records in the left table

```
df_7 = _deepnote_execute_sql('select * \nfrom employee as e\nright join department as d on e.dep_id=d.dep_id', 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_7
```

```
df_8 = _deepnote_execute_sql('select * \nfrom department as d\nright join employee as e on d.dep_id=e.dep_id', 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_8
```

### FULL JOIN 

Returns all records of the tables to be merged

```
df_9 = _deepnote_execute_sql('select * \nfrom department as d\nfull join employee as e on d.dep_id=e.dep_id', 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_9
```

### CROSS JOIN 

returns all possible combinations of the rows of both tables

```
df_11 = _deepnote_execute_sql('select * \nfrom department as d\ncross join employee as e ', 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_11
```

### SELF JOIN 

combines tables with itself

```
df_11 = _deepnote_execute_sql('select d1.* \nfrom department as d1 \ninner join department as d2 ON d1.dep_id=d2.dep_id', 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_11
```
